# 보고서

from sklearn.cluster import KMeans 는 주어진 데이터를 k개의 클러스터로 묶어 각 클러스터와 거리 차이의 분산을 최소화하는 방식으로 동작하는 알고리즘을 불러오는 것이다. 이를 이용해 클러스터의 개수에 따라 달라지는 kiner값으로 기울기가 어느정도 완만해지기 시작하는 지점을 볼 수 있는데 그 지점을 기준으로 언더피팅과 오버피팅의 정도를 파악할 수 있다. 또한 이 이미지에서 알 수 있다시피 그 지점에서는 이미지를 어느정도 파악할 수 있는 지점이라고 생각할 수 있다.
